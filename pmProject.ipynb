```
commentTable = pd.read_sql_query("SELECT * FROM t_comments", conn)
commentTable
```

```
commentData = commentTable['comment']
commentData
```

**Preprocessing**

```
""" Preprocessing comment """

# Change to lowercase
commentData = commentData.str.lower()
# remove number
commentData = commentData.str.replace(r'\d+', '')
# remove puctuation
commentData = commentData.str.replace(r'[^\w\s]', '')
# remove whitespace
commentData = [comment.strip() for comment in commentData]
commentData
```

```
commentDF = pd.DataFrame({
    'comment' : [comment for comment in commentData]
})
```

```
commentDF['tokenizedComment'] = [comment.split() for comment in commentDF['comment']]
commentDF
```
